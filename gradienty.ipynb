# Przykładowy kąt dla wektora skierowanego do góry i wektora słońca 45 (-45 stopni) stopni powinno dać 135 stopni lub 225 stopni

```python
# Przykładowe wektory
A = np.array([-np.sqrt(2)/2, 0, -np.sqrt(2)/2])
B = np.array([0, 0, 1])

# Oblicz iloczyn skalarny
dot_product = np.dot(A, B)

# Oblicz normy (długości) wektorów
norm_A = np.linalg.norm(A)
norm_B = np.linalg.norm(B)

# Oblicz kąt między wektorami w radianach
angle_rad = np.arccos(dot_product / (norm_A * norm_B))

# Przelicz na stopnie
angle_deg = np.degrees(angle_rad)
print(angle_rad, angle_deg)
```

Założenia:
1. Wyliczenia kątu normalnego, prostopadłego do płaszczyzny, należy prowadzić w sposób następujący. Dokonujemy iloczynu wektorowego, z wektorów pomiędzy punktem środkowym a dolnym(S) - wektor AB, oraz pomiędzy punktem środkowym a prawym(E)- wektor AC, w przypadku ostatniej kolumny iloczyn wektorowy pomiędzy środkowym a prawym punktem(W)-wektor AB, oraz pomiędzy punktem środkowym, a dolnym(S)- wektor AC, w przypadku ostatniego wersu to środek i (E) - wektor AB i środek z (N) - wektor AC. Ostatni punkt w prawym dolnym rogu mapy to wektory pomiędzy środkiem a (N) jako AB, oraz pomiędzy środkiem a (W) jako AC. Powstały wektor N należy znormalizować, czyli podzielić przez jego długość:

a. AB=B-A, AC=C-A, N=ABxAC, Nlen=sqrt(nx^2+ny^2+nz^2), Nnorm=N/Nlen

2. Mamy wyliczone wektory prostopadłe do trójkątów, które są w dodatku skierowane do góry(po to były oznaczenia AB i AC, ponieważ mnożenie wektorów nie jest przemienne), potrzeba wyliczyć wektor odbity (R) względem wektora normalnego Nnorm dla wektora promieni słonecznych (L):

b. z biblioteki numpy jako np- R = L - 2 * np.dot(L, N) * N.

3. Najbardziej optymalny wektor dla  obliczeń jest skierowany do góry, jeżeli kąt jest równy 0, to punkt powinien być biały, jeżeli kąt jest większy od 90 stopni lub mniejszy od -90 stopni, to punkt powinien być czarny. Trzeba także pamiętać, że może nie istnieć wektor odbity, bo ściana będzie równoległa do promieni słonecznych.

c.θ=arccos(RoL/∥R∥⋅∥L∥), gdzie podwójne nawiasy to norma danego wektoru, a o to iloczyn skalarny.
